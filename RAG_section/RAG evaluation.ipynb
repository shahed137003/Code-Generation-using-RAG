````python

import pandas as pd

df = pd.read_parquet("hf://datasets/openai/openai_humaneval/openai_humaneval/test-00000-of-00001.parquet")
df.drop(columns=['test','entry_point'],inplace=True)

df
````

````python
from sentence_transformers import SentenceTransformer
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

def embedding_process(input_text):
  return embedding_model.encode(input_text)
````

````python

## converting the prompt field into embedding vectors
df['embedded_prompt']=df['prompt'].map(embedding_process)
df
     
````

````python
## setup the chromadb
import chromadb
from chromadb.config import Settings

client = chromadb.Client(Settings())

collection = client.create_collection(name="humaneval_prompts")
````

````python
 ## creating the collection
collection.add(
       documents=df['prompt'].tolist(),              # The original prompts
    ids=df['task_id'].astype(str).tolist(),
    embeddings=df['embedded_prompt'].tolist(),          # The embedding vectors
    metadatas=[{"solution": s} for s in df['canonical_solution']]
)
collection
````

````python
def search(query_embedding_input):
  return collection.query(
    query_embeddings=query_embedding_input,
    n_results=3)
  
````

````python

from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model_id = "meta-llama/Llama-2-7b-chat-hf"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map={"": device}
)
````

````python
def rag_code_generation(user_prompt):
    """
    RAG pipeline:
    - Embed prompt
    - Retrieve relevant examples
    - Construct formatted prompt
    - Generate new code
    """
    # Step 1: Embed the user prompt
    embedding = embedding_process(user_prompt).tolist()

    # Step 2: Retrieve from vector database
    search_result = search(embedding)
    retrieved_info = search_result['documents'][0]  # List of 3 similar prompts
    solution = search_result['metadatas'][0]        # List of dicts with 'solution'

    # Step 3: Format prompt using retrieved info
    def formatting_prompt(retrieved_info, solution_info):
        context = ""
        for i in range(len(retrieved_info)):
            context += f"### Task {i+1}: {retrieved_info[i]}\n"
            context += f"### Solution {i+1}:\n{solution_info[i]['solution']}\n\n"
        return context + f"### New Task: {user_prompt}\n### Solution:\n"

    prompt = formatting_prompt(retrieved_info, solution)

    # Step 4: Tokenize input
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True)

    # Step 5: Move to device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    inputs = {k: v.to(device) for k, v in inputs.items()}
    model.to(device)

    # Step 6: Generate output
    outputs = model.generate(
        **inputs,
        max_new_tokens=200,
        do_sample=True,
        temperature=0.7,
        top_p=0.9,
        repetition_penalty=1.1
    )

    full_output = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Step 7: Extract the generated code only
    split_marker = f"### New Task: {user_prompt}\n### Solution:\n"
    if split_marker in full_output:
        generated_code = full_output.split(split_marker)[-1].strip()
    else:
        generated_code = full_output.strip()

    return generated_code

````

````python
import re
import traceback
from typing import List
from tqdm import tqdm

# --------------------------------------
# Extracts clean Python code from LLM output
# --------------------------------------
def extract_code_with_imports(text: str) -> str:
    """
    Removes triple backticks and leading/trailing whitespace.
    Keeps valid Python code only.
    """
    # Remove all ```python or ``` markers
    text = re.sub(r"```(?:python)?", "", text)
    text = text.replace("```", "")
    return text.strip()

# --------------------------------------
# Renames the first function in code to the expected one (matches test cases)
# --------------------------------------
def rename_function(code_str: str, expected_name: str) -> str:
    return re.sub(r"def\s+\w+", f"def {expected_name}", code_str, count=1)

# --------------------------------------
# Runs test cases safely on the generated code
# --------------------------------------
def run_tests(code_str: str, tests: List[str]) -> bool:
    local_env = {}
    try:
        exec(code_str, local_env)
        for test_case in tests:
            exec(test_case, local_env)
        print(" All tests passed!")
        return True
    except Exception:
        print(" Test failed:")
        traceback.print_exc()
        return False

# --------------------------------------
# Full RAG Evaluation Loop
# --------------------------------------
def evaluate_rag(sample_data, rag_code_generation):
    total = len(sample_data)
    passed_count = 0

    for sample in tqdm(sample_data, desc="ðŸ” Evaluating RAG"):
        task = sample["text"]
        reference_code = sample.get("code", "")
        test_cases = sample["test_list"]

        # Try to extract the expected function name from the first test case
        match = re.search(r"assert\s+([a-zA-Z_][a-zA-Z0-9_]*)\(", test_cases[0])
        expected_name = match.group(1) if match else None

        print(f"\n Task: {task}")

        # Step 1: Generate code using RAG system
        generated_text = rag_code_generation(task)
        print(" Raw Generated Text:\n", generated_text)

        # Step 2: Clean and extract valid Python code
        code_generated = extract_code_with_imports(generated_text)

        # Step 3: Rename function if test expects a specific name
        if expected_name:
            code_generated = rename_function(code_generated, expected_name)

        print(" Final Code:\n", code_generated)

        # Step 4: Run the test cases
        passed = run_tests(code_generated, test_cases)
        print(" Test Result:", passed)
        print("â”€" * 60)
        passed_count += int(passed)




 

evaluate_rag(sample_data, rag_code_generation)
````
